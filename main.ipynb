## Importing Libraries and required csv data for model training

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pandas import DataFrame, Series
df = pd.read_csv('financialMarketData.csv')
pd.set_option('display.max_columns', None)
df = df.drop('Data', axis=1)
# df = df.drop([i for i in range(200)])
# df = df.drop(df.columns[10:], axis=1)
df
```

### Categorizing Training and Testing Dataset

```python
X_featureColumn = df[df.columns[1:]]
Y_targetColumn = df[df.columns[0]]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_featureColumn, Y_targetColumn, random_state=17, test_size=0.2)
```

## RandomForest training method

```python
forward_fill_features = ['XAU BGNL', 'BDIY', 'CRY', 'DXY', 'Cl1', 'VIX', 'MXUS', 'MXEU', 'MXJP', 'MXBR', 'MXRU', 'MXIN', 'MXCN', 'GBP']
knn_impute_features = ['USGG30YR', 'GT10', 'USGG2YR', 'USGG3M', 'US0001M', 'GTDEM30Y', 'GTDEM10Y', 'GTDEM2Y',
                           'GTITL30YR', 'GTITL10YR', 'GTITL2YR', 'GTJPY30YR', 'GTJPY10YR', 'GTJPY2YR', 'GTGBP30Y', 
                           'GTGBP20Y', 'GTGBP2Y', 'LUMSTRUU', 'LMBITR', 'LUACTRUU', 'LF98TRUU', 'LG30TRUU', 'LP01TREU', 
                           'EMUSTRUU', 'LF94TRUU']
median_impute_features = ['ECSURPUS', 'EONIA']
expected_features = ['XAU BGNL', 'ECSURPUS', 'BDIY', 'CRY', 'DXY', 'JPY', 'GBP', 'Cl1',
       'VIX', 'USGG30YR', 'GT10', 'USGG2YR', 'USGG3M', 'US0001M', 'GTDEM30Y',
       'GTDEM10Y', 'GTDEM2Y', 'EONIA', 'GTITL30YR', 'GTITL10YR', 'GTITL2YR',
       'GTJPY30YR', 'GTJPY10YR', 'GTJPY2YR', 'GTGBP30Y', 'GTGBP20Y', 'GTGBP2Y',
       'LUMSTRUU', 'LMBITR', 'LUACTRUU', 'LF98TRUU', 'LG30TRUU', 'LP01TREU',
       'EMUSTRUU', 'LF94TRUU', 'MXUS', 'MXEU', 'MXJP', 'MXBR', 'MXRU', 'MXIN',
       'MXCN']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.impute import KNNImputer, SimpleImputer

def calculateRandomForest(minNumberSplit:int = 2, maxDepth:int=None, classWeight=None) -> tuple[RandomForestClassifier, dict]:
  rf = RandomForestClassifier(
      random_state=42,
      min_samples_split = minNumberSplit,
      class_weight = classWeight,
      max_depth = maxDepth,
  )
  
  knn_imputer = KNNImputer(n_neighbors=5)
  median_imputer = SimpleImputer(strategy='median')
  ffil_median_imputer = SimpleImputer(strategy='median')

  X_train[knn_impute_features] = knn_imputer.fit_transform(X_train[knn_impute_features])
  X_train[median_impute_features] = median_imputer.fit_transform(X_train[median_impute_features])
  X_train[forward_fill_features] = ffil_median_imputer.fit_transform(X_train[forward_fill_features])
  
  
  X_test[knn_impute_features] = knn_imputer.transform(X_test[knn_impute_features])
  X_test[median_impute_features] = median_imputer.transform(X_test[median_impute_features])
  
  X_test[forward_fill_features] = X_test[forward_fill_features].ffill().bfill().interpolate()
  for col in forward_fill_features:
    if X_test[col].isna().all():  # If the column is completely empty
      X_test[col] = ffil_median_imputer.transform(X_test[[col]])

  rf.fit(X_train, y_train)
  Y_pred = rf.predict(X_test)
  accuracy = accuracy_score(y_test, Y_pred)
  print("Accuracy:", accuracy)
  print("Confusion Matrix:")
  print(confusion_matrix(y_test, Y_pred))
  print("Classification Report:")
  print(classification_report(y_test, Y_pred))
  return rf, {"knn_imputer": knn_imputer, "median_imputer": median_imputer, "ffil_median_imputer": ffil_median_imputer}
```

```python
import joblib
rfModel, hmOfImputers = calculateRandomForest(classWeight={0:1, 1:1})
joblib.dump(rfModel, "./model/random_forest.joblib")
joblib.dump(hmOfImputers["knn_imputer"], './model/imputers/knn_imputer.joblib')
joblib.dump(hmOfImputers["median_imputer"], './model/imputers/median_imputer.joblib')
joblib.dump(hmOfImputers["ffil_median_imputer"], './model/imputers/ffil_median_imputer.joblib')
```

```python
loaded_rf = joblib.load("./model/random_forest.joblib")
knn_imputer = joblib.load("./model/imputers/knn_imputer.joblib")
median_imputer = joblib.load("./model/imputers/median_imputer.joblib")
# Testing the Data

X_test_copy = imputeData(X_test.copy())

Y_pred = loaded_rf.predict(X_test_copy)
accuracy = accuracy_score(y_test, Y_pred)
print("Accuracy:", accuracy)
print("Confusion Matrix:")
print(confusion_matrix(y_test, Y_pred))
print("Classification Report:")
print(classification_report(y_test, Y_pred))
```
